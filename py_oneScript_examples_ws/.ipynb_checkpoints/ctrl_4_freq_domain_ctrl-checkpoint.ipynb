# Frequency Domain Control

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# System parameters
omega_n = 1.0  # Natural frequency
zeta = 0.1    # Damping ratio

# Transfer function for the second-order system
numerator = [omega_n**2]
denominator = [1, 2 * zeta * omega_n, omega_n**2]
system_tf = ctrl.TransferFunction(numerator, denominator)

# Bode plot
ctrl.bode_plot(system_tf)
plt.show()
```
