# Stability Analysis
### Plots used for Stability Analysis
1. Pole-Zero Map
2. Bode Plot
3. Nyquist Plot
4. Root Locus Plot

### Pole-Zero Map
- **Purpose:** Visualize the location of poles and zeros in the complex plane.
- **Interpretation:** A system is stable if all poles are located in the left-half plane (LHP) of the complex plane.

```python
import control as ctrl
import matplotlib.pyplot as plt

# Define the transfer function
numerator = [1]
denominator = [1, 1]
sys = ctrl.TransferFunction(numerator, denominator)

# Plot the poles
ctrl.pzmap(sys, grid=True)
plt.title('Pole-Zero Map')
plt.show()

# Check stability
is_stable, _ = ctrl.stability.margins(sys)
if is_stable:
    print('The system is stable.')
else:
    print('The system is unstable.')
```

### Bode Plot
- **Purpose:** Display the frequency response of a system.
- **Interpretation:** Gain and phase margins can be read directly from the Bode plot to assess stability margins.

```python
import control as ctrl
import matplotlib.pyplot as plt

sys = ctrl.TransferFunction([1], [1, 2, 1])  # Example transfer function
mag, phase, omega = ctrl.bode(sys)
plt.show()
```

### Nyquist Plot
- **Purpose:** Plot the frequency response in polar coordinates.
- **Interpretation:** Stability can be assessed by examining the encirclement of the critical point (-1, 0) in the Nyquist plot.

```python
import control as ctrl
import matplotlib.pyplot as plt

sys = ctrl.TransferFunction([1], [1, 2, 1])  # Example transfer function
_, _ = ctrl.nyquist(sys)
plt.show()
```

### Root Locus Plot
- **Purpose:** Visualize the movement of poles as a parameter (usually gain) is varied.
- **Interpretation:** Stability and transient response characteristics can be analyzed by examining the root locus plot.

```python
import control as ctrl
import matplotlib.pyplot as plt

sys = ctrl.TransferFunction([1], [1, 2, 1])  # Example transfer function
_, _ = ctrl.rlocus(sys)
plt.show()
```
