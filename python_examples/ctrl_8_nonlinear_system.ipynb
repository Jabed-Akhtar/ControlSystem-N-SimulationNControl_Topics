# Non-Linear Systems
Non-linear systems are systems in which the relationship between the input and output is not proportional or additive. In control systems, non-linearities can arise due to various reasons such as saturation, dead zones, backlash, and other complex dynamics.  

A **linear map** (or **linear function**) f(x) is one which satisfies both of the following properties:  
- Additivity or superposition principle: *f(x+y) = f(x) + f(y)*  
- Homogeneity: *f(ax) = af(x)*  

A **non-linear function** can be knows as the function which doesn't satisty the above properties. Furthermore, Nonlinear algebraic equations, which are also called polynomial equations, are defined by equating polynomials (of degree greater than one) to zero. For example: x^2 + x - 1 = 0. [Source: https://en.wikipedia.org/wiki/Nonlinear_system]

### Example#1: Path Planning
In path planning, a common non-linear element is the curvature of the path. Let's consider a simple path represented by a sinusoidal function:  
<img src="notebook_imgs/ctrl_nonlinear_pathplanning_example.png" alt="Example Image" width="200"/>  
Here, A, B, and C are parameters determining the amplitude, frequency, and phase of the path.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the path parameters
A = 1.0  # Amplitude
B = 1.0  # Frequency
C = 0.0  # Phase

# Generate time values
t = np.linspace(0, 2*np.pi, 1000)

# Generate the path
y = A * np.sin(B*t + C)

# Plot the path
plt.plot(t, y)
plt.title('Path Planning: Sinusoidal Path')
plt.xlabel('Time')
plt.ylabel('Path')
plt.grid(True)
plt.show()
```

### Example#2: Longitudinal Trajectory
In longitudinal trajectory, non-linearities can arise due to factors like tire slip. A simple non-linear model for longitudinal dynamics is given by:  
<img src="notebook_imgs/ctrl_nonlinear_lateral_example.png" alt="Example Image" width="150"/>  
Here, v is the lateral velocity, r is the yaw rate, F_y is the lateral force, m is the mass, I_z is the moment of inertia, and a and b are the distances from the center of mass to the front and rear axles, respectively.

```python
# Define system parameters
m = 1500.0  # Mass (kg)
Iz = 3000.0 # Moment of inertia (kg.m^2)
a = 1.0     # Distance from CG to front axle (m)
b = 1.5     # Distance from CG to rear axle (m)

# Generate lateral force values
Fy = np.linspace(-5000, 5000, 100)

# Compute lateral velocity and yaw rate
v_dot = Fy / m
r_dot = (a*Fy - b*Fy) / Iz

# Plot lateral velocity and yaw rate vs. lateral force
plt.plot(Fy, v_dot, label='Lateral Velocity')
plt.plot(Fy, r_dot, label='Yaw Rate')
plt.title('Lateral Dynamics: Velocity and Yaw Rate vs. Lateral Force')
plt.xlabel('Lateral Force (N)')
plt.ylabel('Velocity and Yaw Rate')
plt.legend()
plt.grid(True)
plt.show()
```

This represents a non-linear relationship between lateral force and both lateral velocity and yaw rate in lateral dynamics.
