### Example#1: A Second-Order System
Now, let's implement this system representation in Python using the control package. If you don't have the control package installed, you can install it using pip install control.

```python
import control as ctrl
import matplotlib.pyplot as plt

# System parameters
omega_n = 2.0  # Natural frequency
zeta = 0.7     # Damping ratio

# Transfer function
numerator = [omega_n**2]
denominator = [1, 2 * zeta * omega_n, omega_n**2]
sys = ctrl.TransferFunction(numerator, denominator)

# Display the system transfer function
print("System Transfer Function:")
print(sys)

# Step response
time, response = ctrl.step_response(sys)

# Plot the step response
plt.plot(time, response)
plt.title('Step Response of the Second-Order System')
plt.xlabel('Time')
plt.ylabel('Response')
plt.grid(True)
plt.show()
```

### Example#2: Longitudinal Trajectory
Longitudinal trajectory control involves controlling the vehicle's speed. A simple representation can be a first-order system.  
Formula for a first-order system:  
<img src="notebook_imgs/ctrl_sys_repr_long_trj_example.png" alt="Example Image" width="180"/>  

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# System parameters
a = 0.1  # Gain
v_desired = 10.0  # Desired velocity

# System dynamics
num = [a]
den = [1, a]

system = ctrl.TransferFunction(num, den)

# Time vector
time = np.linspace(0, 100, 100)

# Step response
_, response = ctrl.step_response(system, time)

plt.plot(time, response)
plt.title('Longitudinal Trajectory Control')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.grid()
plt.show()
```

### Example#3: Lateral Trajectory
Lateral trajectory control involves steering the vehicle along the planned path. A simple representation can be a proportional controller.  
Formula for a propertional controller:  
<img src="notebook_imgs/ctrl_sys_repr_lat_trj_example.png" alt="Example Image" width="200"/>

```python
import matplotlib.pyplot as plt
import numpy as np

# Controller parameters
K_p = 0.1  # Proportional gain

# Error signal (difference between desired and actual lateral position)
lateral_error = np.random.rand(100)

# Steering command
steering_command = K_p * (lateral_error - np.roll(lateral_error, 1))

plt.plot(steering_command)
plt.title('Lateral Trajectory Control')
plt.xlabel('Time')
plt.ylabel('Steering Command')
plt.show()
```
